```python
def create_prompt_formats(sample):
    """
    Format various fields of the sample ('question_cand_answer', 'justification_judge')
    Then concatenate them using two newline characters
    :param sample: Sample dictionary containing interview data.
    """
    # Introductory prompt describing the task for the AI assistant
    intro_prompt = """
    <|begin_of_text|><|start_header_id|>system<|end_header_id|>
    You are a highly capable and unbiased AI assistant tasked with evaluating a candidate's technical competence based on their responses during an interview. Your role is to provide an accurate and fair assessment, focusing on the candidate's understanding, problem-solving abilities, and depth of knowledge in relevant technical domains. You should consider the clarity, correctness, and thoroughness of the candidate's explanations.<|eot_id|>
    """

    # Instruction key describing what needs to be done
    INSTRUCTION_KEY = """
    <|start_header_id|>user<|end_header_id|>
    Please analyze the following exchange between an interviewer and a candidate:

    Based on the candidate's response, provide a detailed judgment focusing on the following aspects:
    1. **Technical Knowledge**: Evaluate the depth and accuracy of the candidate's technical understanding.
    2. **Problem-Solving Skills**: Assess the candidate's ability to address and solve potential challenges or issues.
    3. **Clarity and Communication**: Consider how clearly and effectively the candidate communicated their ideas and solutions.
    4. **Relevance and Applicability**: Judge how relevant and applicable the candidate's response is to the question asked and to the practical scenarios they may encounter in the role.
    What is the judgement you can give me on this person based on the answers given by the person for the questions asked
    <|eot_id|>
    """

    # Response key where the AI assistant is expected to give a judgment
    RESPONSE_KEY = """

    <|start_header_id|>assistant<|end_header_id|>
    Based on the above conversation and the rules specified for the judgement, here is my detailed judgment:
    """

    END_KEY = "### End"

    # Split the interview question and answer
    split_text = sample["question_cand_answer"].split("\n\nInterviewee: ")

    if len(split_text) == 2:
        question = split_text[0].replace("Interviewer: ", "").strip()
        answer = split_text[1].strip()
    else:
        question = sample["question_cand_answer"].strip()
        answer = ""

    # Formatting the input context with the question and answer
    input_context = f"### Question : {question}\n\n### Answer : {answer}"

    # Formatting the response with the judgment
    response = f"{RESPONSE_KEY}\n### Judgment : {sample['justification_judge']}"

    # Assemble all parts together
    parts = [part.strip() for part in [intro_prompt, INSTRUCTION_KEY, input_context, response, END_KEY] if part]

    # Concatenate with two newlines
    formatted_prompt = "\n\n".join(parts)

    # Assign the formatted prompt to the sample's text field
    sample["text"] = formatted_prompt

    return sample
```

```python
def get_max_length(model):
    conf = model.config
    max_length = None
    for length_setting in ["n_positions", "max_position_embeddings", "seq_length"]:
        max_length = getattr(model.config, length_setting, None)
        if max_length:
            print(f"Found max lenth: {max_length}")
            break
    if not max_length:
        max_length = 1024
        print(f"Using default max length: {max_length}")
    return max_length


def preprocess_batch(batch, tokenizer, max_length):
    """
    Tokenizing a batch
    """
    return tokenizer(
        batch["text"],
        max_length=max_length,
        truncation=True,
    )
```

```python
from functools import partial

# SOURCE https://github.com/databrickslabs/dolly/blob/master/training/trainer.py
def preprocess_dataset(tokenizer: AutoTokenizer, max_length: int,seed, dataset):
    """Format & tokenize it so it is ready for training
    :param tokenizer (AutoTokenizer): Model Tokenizer
    :param max_length (int): Maximum number of tokens to emit from tokenizer
    """

    # Add prompt to each sample
    print("Preprocessing dataset...")
    dataset = dataset.map(create_prompt_formats)#, batched=True)

    _preprocessing_function = partial(preprocess_batch, max_length=max_length, tokenizer=tokenizer)
    dataset = dataset.map(
        _preprocessing_function,
        batched=True,
        #remove_columns=['id', 'topic', 'dialogue', 'summary'],
    )

    # Filter out samples that have input_ids exceeding max_length
    # dataset = dataset.filter(lambda sample: len(sample["input_ids"]) < max_length)

    # Shuffle dataset
    dataset = dataset.shuffle(seed=seed)

    return dataset
```

```python
# ## Pre-process dataset
max_length = get_max_length(original_model)
print(max_length)

train_dataset = preprocess_dataset(tokenizer, max_length,seed, dataset['train'])
eval_dataset = preprocess_dataset(tokenizer, max_length,seed, dataset['test'])
```

```python
def print_number_of_trainable_model_parameters(model):
    trainable_model_params = 0
    all_model_params = 0
    for _, param in model.named_parameters():
        all_model_params += param.numel()
        if param.requires_grad:
            trainable_model_params += param.numel()
    return f"trainable model parameters: {trainable_model_params}\nall model parameters: {all_model_params}\npercentage of trainable model parameters: {100 * trainable_model_params / all_model_params:.2f}%"

print(print_number_of_trainable_model_parameters(original_model))
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

config = LoraConfig(
    r=32, #Rank
    lora_alpha=32,
    target_modules=[
        'q_proj',
        'k_proj',
        'v_proj',
        'dense'
    ],
    bias="none",
    lora_dropout=0.05,  # Conventional
    task_type="CAUSAL_LM",
)

# 1 - Enabling gradient checkpointing to reduce memory usage during fine-tuning
original_model.gradient_checkpointing_enable()

# 2 - Using the prepare_model_for_kbit_training method from PEFT
original_model = prepare_model_for_kbit_training(original_model)

peft_model = get_peft_model(original_model, config)
```

```python
output_dir = './peft-dialogue-summary-training/final-checkpoint'
import transformers

peft_training_args = TrainingArguments(
    output_dir=output_dir,
    warmup_steps=1,
    per_device_train_batch_size=1,  # Reduce batch size
    gradient_accumulation_steps=12,
    max_steps=1,
    learning_rate=2e-4,
    optim="paged_adamw_8bit",
    logging_steps=1,
    logging_dir="./logs",
    save_strategy="steps",
    save_steps=1,
    evaluation_strategy="steps",
    eval_steps=1,
    do_eval=True,
    gradient_checkpointing=True,
    report_to="none",
    overwrite_output_dir=True,
    group_by_length=True,
    fp16=True,  # Enable mixed precision training
)

peft_model.config.use_cache = False

peft_trainer = transformers.Trainer(
    model=peft_model,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    args=peft_training_args,
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)

# Optionally clear the CUDA cache
import torch
torch.cuda.empty_cache()

# Train the model
peft_trainer.train()
```
